```python
import tensorflow as tf
train_ds = tf.keras.utils.image_dataset_from_directory('datasets', image_size=(1080, 1920))
train_ds
```

```python
import matplotlib.pyplot as plt

class_names = train_ds.class_names
plt.figure(figsize=(20, 20))
for images, labels in train_ds.take(1):
    image = images[0].numpy().astype("uint8")
    plt.imshow(image)
    plt.title(class_names[labels[0]])
    plt.axis("off")
```

```python
box = tf.image.crop_to_bounding_box(image, 160, 310, 760, 1060)
box
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(box)
plt.axis("off")
```

```python
multiple = .5
size = (int(box.shape[0] * multiple), int(box.shape[1] * multiple))
size
```

```python
resize = tf.image.resize(box, size).numpy().astype("uint8")
resize
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(resize)
plt.axis("off")
```
